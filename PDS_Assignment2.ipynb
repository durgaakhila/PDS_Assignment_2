# Steps that followed in this code as per the problem statement in this assignment

1. Before handling the missing values,I removed units to make them float type values. Because, most of the columns are object type values i.e., float value is concatenated with string values called units.
2. After removing units I handled the missing values
3. Then, I converted the fuel_type and transmission column values into one hot encoding
4. After that instead using R mutate function to create current_age feature for each car using datetime library which is an python equivalent to R. And please find the R mutate code in comments.

```python
import pandas as pd
```

```python
file_path = "/content/train.csv"
```

```python
data = pd.read_csv(file_path)
```

```python
data.head(10)
```

```python
data.info()
```

```python
# In above info we go to see missing values in Mileage,Engine,Power,Seats,New_Price
# Interstingly, Engine is object ex:1582 CC
# Mileage is also an object ex : 19.67 kmpl
# same as all others are objects what I can do is I remove their units first then I handle the missing values

# Remove 'kmpl' and 'km/kg' units from 'Mileage' column
data['Mileage'] = data['Mileage'].astype(str).str.replace(' kmpl', '').str.replace(' km/kg', '').astype(float)

# Remove ' CC' from 'Engine' column
data['Engine'] = data['Engine'].astype(str).str.replace(' CC', '').astype(float)

# Remove ' bhp' from 'Power' column
data['Power'] = data['Power'].astype(str).str.replace(' bhp', '').astype(float)

# Remove ' Lakh' and 'Cr' from 'New_price' column
data['New_Price'] = data['New_Price'].astype(str).str.replace(' Lakh', '').str.replace(' Cr', '').astype(float)
```

```python
data.head()
```

```python
missing_values = data.isnull().sum()
```

```python
print(missing_values)
```

```python
# Here I found the missing values in following attributes and they're Engine, Mileage, Power, seats, New_price
# And I use following Impute strategy for them that is for mileage and engine I drop missing values in rows
# Then For seats i handle missing values with mean
# Then For Power i handle missing values with median
# Then For New_Price i handle missing value with mode

data.dropna(subset=['Mileage', 'Engine'], inplace=True)

data['Seats'].fillna(data['Seats'].mean(), inplace=True)

data['Power'].fillna(data['Power'].mean(), inplace=True)

data['New_Price'].fillna(data['New_Price'].mean(), inplace=True)
```

```python
data.head(10)
```

```python
missing_values= data.isnull().sum()
```

```python
print(missing_values)
```

```python
# Now we have removed the units as well as handled the missing values
# As of now Fuel_Type has value of Automatic/Manual which is an categorical value
# And Transmission has value of Diesel/Petrol which is also an categorical value

# So using one hot encoding I convert both variables as numeric in 0/1 using pd.dummies function which us one hot encoding inherently

# Use pd.get_dummies() to one-hot encode the categorical columns into binary(0/1) columns
data = pd.get_dummies(data, columns=['Fuel_Type', 'Transmission'])
```

```python
data.head()
```

```python
# I am creating a new feature for current age of car in python (Since in question it mentioned to write in R but i am providing in python equivalent)

from datetime import datetime

current_year = datetime.now().year
data['Current_Age'] = current_year - data['Year']
```
